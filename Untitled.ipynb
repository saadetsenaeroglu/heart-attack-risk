```python
import pandas as pd 
import numpy as np

from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score
# Gerekli kütüphaneleri ekledim ve o kütüphanelerden kullanacağım fonksiyonları içeri çektim.

df = pd.read_csv("heart.csv")
df.head()
# Datasette hata olup olmadığını kontrol ettim.
```

```python
df.info()
#Makine öğreimi için veri tipi string olanları encode edebilmek için önce hangi sütunların string olduğuna baktım.
#Aynı zamanda Datasetimi tanıdım.
```

```python
# Encode işlemlerinde zorlandığım için bu adımlarda yapay zekadan yardım aldım.
# 'Gender','prevalentStroke','Heart_ stroke' sütunlarını encode ettim.
df['Gender'].unique()
df['Gender'].value_counts(dropna=False)
```

```python
df['Gender'] = (
    df['Gender'].astype(str).str.strip().str.lower()
      .map({'male': 0, 'female': 1})
)

print(df['Gender'].value_counts(dropna=False))
```

```python
stroke_norm = df['prevalentStroke'].astype(str).str.strip().str.lower()
mapping = {
     'yes': 1, 'no': 0
}

df['prevalentStroke'] = stroke_norm.map(mapping)
print(df['prevalentStroke'].value_counts(dropna=False))
```

```python
stroke_norm = df['Heart_ stroke'].astype(str).str.strip().str.lower()
mapping = {
  'yes': 1,'no': 0
}

df['Heart_ stroke'] = stroke_norm.map(mapping)
print(df['Heart_ stroke'].value_counts(dropna=False))
```

```python
df.head()
```

```python
# Önemsiz kategorileri ve çok fazla eksik veri olan kategorileri silerek veri setimi düzenledim.
```

```python
df = df.drop(columns=["glucose"], errors="ignore")
```

```python
df = df.drop(columns=["cigsPerDay"], errors="ignore")
```

```python
df = df.drop(columns=["education"], errors="ignore")
```

```python
df.isna().sum()
```

```python
# Makine öğrenmesi için elimdeki veri setini X(özellikler) ve y (sınıf) olarak ayırdım.
X = df[["Gender","age","currentSmoker","BPMeds","prevalentStroke","prevalentHyp","diabetes","totChol","sysBP","diaBP","BMI","heartRate"]]                                    
y = df["Heart_ stroke"] 

print("'X'in ilk 5 satırı:")
print(X.head())
print()
print("'y'nin ilk 5 satırı:")
print(y.head())
```

```python
X = X.fillna(X.median())
#eksik verileri doldurmayı biilmediğim için yapay zeka uygulamasından yardım alarak bu kod satırını oluşturdum.
```

```python
# Eğitim ve test verilerini ayırarak test için ağırlık oluşturdum.
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.4,
    random_state=42 
)
print("Eğitim veri boyutu:", X_train.shape, y_train.shape)
print()
print("Test veri boyutu:", X_test.shape, y_test.shape)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=10)
knn_model.fit(X_train, y_train)
# Model burada X_train ve y_traindeki verileri hafızasında tutuyor.
```

```python
y_pred = knn_model.predict(X_test)
print(y_pred[:10])
print(y_test.values[:10])
```

```python
# Son olarak eğittiğim modelin doğruluk değerine bakıyorum.
acs = accuracy_score(y_test, y_pred)
print("modelin doğruluğu:", acs)

sonuclar = X_test.copy()
sonuclar["Gerçek (Heart_ stroke)"] = y_test
sonuclar["Tahmin (Heart_ stroke)"] = y_pred
```
